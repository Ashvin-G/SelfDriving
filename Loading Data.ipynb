```python
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
import random
import pickle
```

```python
DATADIR = "..\SelfDriving_Dataset"
CATEGORIES = ["A", "W", "D"]
```

```python
training_data = []

def create_training_data():
    for category in CATEGORIES:
        path = os.path.join(DATADIR, category)
        class_num = CATEGORIES.index(category)
        for img in os.listdir(path):
            img_path = os.path.join(path, img)
            img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
            img = cv2.resize(img, (100, 100))
            training_data.append([img, class_num])
            
            
    
```

```python
create_training_data()
```

```python
A = []
W = []
D = []

for img, label in training_data:
    if label == 0:
        A.append([img, label])
    elif label == 1:
        W.append([img, label])
    else:
        D.append([img, label])
```

```python
print(len(A), len(W), len(D))
```

```python
straight = W[:len(A)]
right = D[:len(A)]
left = A
```

```python
print(len(straight), len(right), len(left))
```

```python
training_data = []

for data in straight:
    training_data.append(data)
for data in right:
    training_data.append(data)
for data in left:
    training_data.append(data)
```

```python
left_counter = 0
right_counter = 0
straight_counter = 0

for img, label in training_data:
    if label == 0:
        left_counter += 1
    elif label == 1:
        straight_counter += 1
    else:
        right_counter += 1
        
```

```python
random.shuffle(training_data)
```

```python
X = []
y = []
```

```python
for features, label in training_data:
    X.append(features)
    y.append(label)
X = np.array(X).reshape(-1, 100, 100, 1)
```

```python
pickle_out = open("X.pickle", "wb")
pickle.dump(X, pickle_out)
pickle_out.close()

pickle_out = open("y.pickle", "wb")
pickle.dump(y, pickle_out)
pickle_out.close()
```
